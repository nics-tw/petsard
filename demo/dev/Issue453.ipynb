```python
from PETsARD import (
    Loader,
    Processor,
    Synthesizer,
)



cols = [
    'Age', 'Experience', 'Income',
    'ZIP Code', 'Family', 'CCAvg',
    'Education', 'Mortgage',
    'Personal Loan',
    'Securities Account',
    'CD Account',
    'Online', 'CreditCard',
]

load = Loader(
    filepath='benchmark://bank_loan',
    # column_types={
    #     'category': [
    #         'Personal Loan',
    #         'Securities Account',
    #         'CD Account',
    #         'Online',
    #         'CreditCard'
    #     ],
    # }
)
load.load()

proc = Processor(metadata=load.metadata,)
proc.fit(data=load.data)
preproc_data = proc.transform(data=load.data,)

preproc_data = preproc_data[cols]

temp_metadata = load.metadata.metadata
# temp_metadata['global']['row_num'] = n_rows
temp_metadata['global']['col_num'] = len(cols)
temp_metadata['col'] =  {key: temp_metadata['col'][key] for key in cols if key in temp_metadata['col']}
load.metadata.metadata = temp_metadata

syn = Synthesizer(method='sdv-copulagan',)
syn.create(data=preproc_data, metadata=load.metadata.to_sdv())
syn.fit_sample()
```

```python
from PETsARD import (
    Loader,
    Processor,
    Synthesizer,
)



cols = [
    'Age', 'Experience', 'Income',
    'ZIP Code', 'Family', 'CCAvg',
    'Education', 'Mortgage', 'Personal Loan',
    'Securities Account',
    'CD Account',
    'Online', 'CreditCard',
]
n_rows = 500

load = Loader(filepath='benchmark://bank_loan')
load.load()

proc = Processor(metadata=load.metadata,)
proc.fit(data=load.data)
preproc_data = proc.transform(data=load.data,)

preproc_data = preproc_data[cols]

temp_metadata = load.metadata.metadata
temp_metadata['global']['row_num'] = n_rows
temp_metadata['global']['col_num'] = len(cols)
temp_metadata['col'] =  {key: temp_metadata['col'][key] for key in cols if key in temp_metadata['col']}
load.metadata.metadata = temp_metadata

syn = Synthesizer(method='smartnoise-mst',)
syn.create(data=preproc_data, metadata=load.metadata.to_sdv())
syn.fit_sample()
```
